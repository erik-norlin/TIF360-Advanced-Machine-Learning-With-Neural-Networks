```python
import numpy as np
import matplotlib.pyplot as plt

ls = 4

img1 = plt.imread('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/{}_bottle_node/3a_n={}_no.1__PRED.png'.format(ls,ls))
img2 = plt.imread('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/{}_bottle_node/3a_n={}_no.1_TRUE.png'.format(ls,ls))

img3 = plt.imread('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/{}_bottle_node/3a_n={}_no.2__PRED.png'.format(ls,ls))
img4 = plt.imread('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/{}_bottle_node/3a_n={}_no.2_TRUE.png'.format(ls,ls))

img5 = plt.imread('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/{}_bottle_node/3a_n={}_no.3__PRED.png'.format(ls,ls))
img6 = plt.imread('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/{}_bottle_node/3a_n={}_no.3_TRUE.png'.format(ls,ls))

# Merge the images horizontally
merge1 = np.concatenate((img1, img2), axis=1)
merge2 = np.concatenate((img3, img4), axis=1)
merge3 = np.concatenate((img5, img6), axis=1)

# Stack the merged images horizontally
final_merge = np.concatenate((merge1, merge2, merge3), axis=0)

# Set the titles for each column and row
column_titles = ['Pred.', 'True']
row_titles = ['img 1', 'img 2', 'img 3']

# Plot the final image with titles
fig, ax = plt.subplots(figsize=(15, 15))
ax.imshow(final_merge)
ax.set_xticks([])
ax.set_yticks([])
fig.suptitle('{} bottle nodes'.format(ls), fontsize=20)

for spine in ax.spines.values():
    spine.set_visible(False)

for i in range(4):
    ax.text(i*img1.shape[1]+img1.shape[1]/2, -60, column_titles[i], ha='center', fontsize=16)

fig.subplots_adjust(hspace=0.5)

# plt.savefig('/content/drive/MyDrive/Skola/CAS/Advanced Machine Learning/HW_B/3a/merged_loss.png')
plt.show()
```
